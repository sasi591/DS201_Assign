```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math as m
```

```python
data1 = pd.read_excel('pca_data.xlsx')
x1 = data1.iloc[:, :2]
y1 = data1.iloc[:, 2]

data2 = pd.read_excel('lda_data.xlsx')
x2 = data2.iloc[:, :2]
y2 = data2.iloc[:, 2]
```

```python
plt.style.use('seaborn')
fig = plt.figure()
plt.scatter(x1[y1==1]['feature 1'], x1[y1==1]['feature 2'], color='r', alpha=0.5, label='Class-1')
plt.scatter(x1[y1==2]['feature 1'], x1[y1==2]['feature 2'], color='b', alpha=0.5, label='Class-2')
plt.title('Scatter Plot Of PCA data', fontsize=20)
plt.xlabel('feature1', fontsize=15)
plt.ylabel('feature2', fontsize=15)
plt.legend(bbox_to_anchor=(1.25, 0.6), fontsize=15)
plt.show()
```

```python
fig = plt.figure()
plt.scatter(x2[y2==1]['feature 1'], x2[y2==1]['feature 2'], color='r', alpha=0.5, label='Class-1')
plt.scatter(x2[y2==2]['feature 1'], x2[y2==2]['feature 2'], color='b', alpha=0.5, label='Class-2')
plt.title('Scatter Plot Of LDA data', fontsize=20)
plt.xlabel('feature1', fontsize=15)
plt.ylabel('feature2', fontsize=15)
plt.legend(bbox_to_anchor=(1.25, 0.6), fontsize=15)
plt.show()
```

## PCA function

```python
def PCA(x, y, datatype='PCA'):
    cov = np.cov(np.dot(x.T, x))
    evals, evecs = np.linalg.eig(cov)
    
    fig, ax = plt.subplots(1, 2, figsize=(20, 5.5))
    plt.subplots_adjust(wspace = 0.5)
    ax[0].scatter(x[y==1]['feature 1'], x[y==1]['feature 2'], color='r', alpha=0.5, label='Class-1')
    ax[0].scatter(x[y==2]['feature 1'], x[y==2]['feature 2'], color='b', alpha=0.5, label='Class-2')
    title='Scatter Plot Of ' + datatype + ' data'
    ax[0].set_title(title, fontsize=20)
    ax[0].set_xlabel('feature1', fontsize=15)
    ax[0].set_ylabel('feature2', fontsize=15)

    mag = (evecs[0, 1]**2 + evecs[1, 1]**2)**0.5 
    proj = np.dot(x, evecs[:, 1]/mag)
    angle = m.atan(evecs[1, 1]/ evecs[0, 1])
    ax[0].scatter(proj[y==1]*m.cos(angle), proj[y==1]*m.sin(angle), color='g', marker='s', label='ProjectedClass-1')
    ax[0].scatter(proj[y==2]*m.cos(angle), proj[y==2]*m.sin(angle), color='m', marker='s', label='ProjectedClass-2')
    ax[0].legend(bbox_to_anchor=(0.95, 0.7), fontsize=15)

    ax[1].scatter(proj[y==1], y[y==1], color='g', marker='s', alpha=0.5, label='Class-1')
    ax[1].scatter(proj[y==2], y[y==2], color='m', marker='s', alpha=0.5, label='Class-2')
    ax[1].set_title('Seperation of Class using PCA 1D Projection', fontsize=20)
    ax[1].set_xlabel('Projection on principle component', fontsize=15)
    ax[1].set_ylabel('Class', fontsize=15)

    ax[1].legend(bbox_to_anchor=(1.25, 0.6), fontsize=15)
    plt.show()
```

```python
def LDA(x, y, datatype='LDA'):
    m1 = np.mean(x[y==1])
    m2 = np.mean(x[y==2])
    
    s1 = np.cov(x[y==1].T)
    s2 = np.cov(x[y==2].T)
    sw = s1 + s2
    
    sb = np.dot((m1-m2), (m1-m2).T)
    evals, evecs = np.linalg.eig(np.dot(np.linalg.inv(sw), sb))
    
    fig, ax = plt.subplots(1, 2, figsize=(20, 5.5))
    plt.subplots_adjust(wspace = 0.5)
    ax[0].scatter(x[y==1]['feature 1'], x[y==1]['feature 2'], color='r', alpha=0.5, label='Class-1')
    ax[0].scatter(x[y==2]['feature 1'], x[y==2]['feature 2'], color='b', alpha=0.5, label='Class-2')
    title='Scatter Plot Of '+ datatype +' data'
    ax[0].set_title(title, fontsize=20)
    ax[0].set_xlabel('feature1', fontsize=15)
    ax[0].set_ylabel('feature2', fontsize=15)

    mag = (evecs[0, 1]**2 + evecs[1, 1]**2)**0.5 
    proj = np.dot(x, evecs[:, 1]/mag)
    angle = m.atan(evecs[1, 1]/ evecs[0, 1])
    ax[0].scatter(proj[y==1]*m.cos(angle), proj[y==1]*m.sin(angle), color='g', marker='s', label='ProjectedClass-1', alpha=0.7)
    ax[0].scatter(proj[y==2]*m.cos(angle), proj[y==2]*m.sin(angle), color='m', marker='s', label='ProjectedClass-2', alpha=0.7)
    ax[0].legend(bbox_to_anchor=(0.95, 0.7), fontsize=15)

    ax[1].scatter(proj[y==1], y[y==1], color='g', marker='s', alpha=0.5, label='Class-1')
    ax[1].scatter(proj[y==2], y[y==2], color='m', marker='s', alpha=0.5, label='Class-2')
    ax[1].set_title('Seperation of Class using LDA 1D Projection', fontsize=20)
    ax[1].set_xlabel('Projection on principle component', fontsize=15)
    ax[1].set_ylabel('Class', fontsize=15)
```
